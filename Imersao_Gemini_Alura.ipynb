Incorporando o modelo

```python
model = "models/embedding-001"
```

```python


def embed_fn(title, text):
  return  genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python


df["Embeddings"] = df.apply(lambda row: embed_fn(row["Nome"],row["Situação"]), axis=1)
df
```

Criando uma função para buscar informações

```python


def gerar_e_buscar_consulta(consulta,df, model):
 embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]
 produtos_escalares = np.dot(np.stack(df["Embeddings"]),embedding_da_consulta)

 indice = int(np.argmax(produtos_escalares))

 return df.iloc[indice]
```

Buscar Informações

```python
consulta = input("Nome do Desaparecido (digite fim para sair):")

while consulta != "fim":
 response = gerar_e_buscar_consulta(consulta,df, model)
 print("")
 print(response)
 print("")
 consulta = input("Nome do Desaparecido (digite fim para sair):")
```
